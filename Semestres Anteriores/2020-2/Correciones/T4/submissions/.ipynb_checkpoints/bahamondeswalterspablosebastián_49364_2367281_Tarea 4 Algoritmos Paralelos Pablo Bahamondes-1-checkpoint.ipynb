```python
import time as tm
import numpy as np
import pyopencl as cl
np.random.seed = 0
```

# Actividades

## 1: Representación números complejos

Dado el problema es necesario rear números complejos y definir las operaciones básicas para números complejos. No se definirán explicitamente como funciones pero explicaré como trataré con ellas.

Dados $w,z$, dos números complejos tales que $w = w_r + w_i i, z = z_r, z_i$ tenemos:

**Suma**:

\begin{align}
z+w &= (z_r + w_r) + (z_i+w_i) i 
\end{align}

**Potencia**:

\begin{align}
z\cdot z = z²  &= (z_r²-z_i²) + (2 z_r z_i) i 
\end{align}

Teniendo en cuenta la memoria y que no debemos perder el valor de $z_n$ pues al calcular la parte real después tenemos que tener cuidado para no utilizarlo en el calculo de la parte imaginaria.

# 2: Explicación de algoritmo

## Escogiendo puntos c =z

En primer lugar queremos buscar puntos aleatorios dentro de un area conocida, para esto creo que es interesante pensar en los número complejos en la forma de euler:

$$ z = re^{i\phi},$$
donde r es el radio y $\phi$ es el angulo de rotación entre el eje real y z. En este sentido es fácil ver que si $r>2$,
$$z² + z = r²e^{2i\phi} + r e^{i\phi} $$
La diferencia se haría maxima si los angulos de cada parte son opuestos, de otro modo: $e^{2i\phi}= -e^{i\phi}$ por lo que la norma más pequeña posible en el mejor de los casos es:

$$|r²-r|>r$$

En otras palabras, su norma comienza a crecer indefinidamente en cada iteración por lo que tomaremos un cuadrado de lado 4 centrado en el origen (me es más simple que un circulo).

## Explicación de algoritmo
Explcaré a grandes rasgos como funcionan 3 programas que implementaré en el código, para esto primero tenemos que considerar 3 partes, la primera en la que se obtiene la *global_id* y se crean 2 variables temporales para guardar la parte real y compleja de los número, luego una parte iterativa que varía de codigo a codigo y finalmente una parte donde se asigna 1 en la posición *global_id* de la variable *result* para saber cuales fueron los números complejos que estuvieron dentro (luego se puede sumar este vector para saber la cantidad de números complejos). A continuación deetallaré un poco mas acerca del funcionamiento de la parte iterativa de los programas:
  1. **Iters** *(program_iters)*: Este programa simplemente eleva el número al cuadrado y suma el complejo inicial hasta que alcanza el número de iteraciones máxima.
  2. **If Statement** *(program_if_statement)*: Este programa eleva el número al cuadrado y suma el complejo inicial para luego comprobar **en cada iteración** si la norma cuadrada del numero es mayor a 4. En caso de que sea mayor a 4, el núcleo de la GPU que este calculando sobre ese valor deja de hacerlo.
  3. **Long Iters** *(program_long_iters)*: Este programa eleva el número al cuadrado y suma el complejo inicial para luego comprobar **cada 1000 iteraciones** si la norma cuadrada del numero es mayor a 4. En caso de que sea mayor a 4, el núcleo de la GPU que este calculando sobre ese valor deja de hacerlo. La idea de este método es no tener que hacer la comparación en cada iteración pues es un calculo costoso (pensemos que sin comparación son necesarias 7 operaciones pero con el *if* se realizan 4 operaciones más).
 

## Escoger cantidad de iteraciones

Por ahora escogeré un número arbitrario de iteraciones que considero grande, 10000, pero más adelante veremos como se va comportanto con una cantidad de iteraciones variables.

```python
%load_ext pyopencl.ipython_ext

plataforms = cl.get_platforms()
print("Plataformas: ", plataforms)

gpu_devices = plataforms[0].get_devices(cl.device_type.GPU)
print("Devices: ", gpu_devices)

context = cl.Context(gpu_devices)
queue = cl.CommandQueue(context)

program = cl.Program(context, 
"""
__kernel void program_iters(
    __global float *a, __global float *b, __global float *a_iters, __global float *b_iters, __global float *result_iters)
{
  int global_id = get_global_id(0);
    
  float temp_i = 0;
  float temp_r = 0;
  
  for (int iterations = 0; iterations< 10000; iterations++)
  {
    temp_r = a_iters[global_id]*a_iters[global_id] - b_iters[global_id]*b_iters[global_id];
    temp_i = 2*a_iters[global_id]*b_iters[global_id];
    
    a_iters[global_id] = temp_r + a[global_id];
    b_iters[global_id] = temp_i + b[global_id];
  }
  
  if (a_iters[global_id]*a_iters[global_id] + b_iters[global_id]*b_iters[global_id] <=4)
  {
  result_iters[global_id] = 1;
  }
}

__kernel void program_if_statement(
    __global float *a, __global float *b, __global float *a_iters, __global float *b_iters, __global float *result_iters)
{
  int global_id = get_global_id(0);
  
  float temp_i = 0;
  float temp_r = 0;
  
  
  for (int iterations = 0; iterations< 10000; iterations++)
  {
    temp_r = a_iters[global_id]*a_iters[global_id] - b_iters[global_id]*b_iters[global_id];
    temp_i = 2*a_iters[global_id]*b_iters[global_id];
    
    a_iters[global_id] = temp_r + a[global_id];
    b_iters[global_id] = temp_i + b[global_id];
    if (a_iters[global_id]*a_iters[global_id] + b_iters[global_id]*b_iters[global_id] >=4)
      {
        break;
      }
  }
  
  if (a_iters[global_id]*a_iters[global_id] + b_iters[global_id]*b_iters[global_id] <=4)
  {
  result_iters[global_id] = 1;
  }
}

__kernel void program_long_iters(
    __global float *a, __global float *b, __global float *a_iters, __global float *b_iters, __global float *result_iters)
{
  int global_id = get_global_id(0);
  
  float temp_i = 0;
  float temp_r = 0;
  
  for (int long_iter = 0; long_iter <10; long_iter ++)
  {
      for (int iterations = 0; iterations< 1000; iterations++)
      {
        temp_r = a_iters[global_id]*a_iters[global_id] - b_iters[global_id]*b_iters[global_id];
        temp_i = 2*a_iters[global_id]*b_iters[global_id];

        a_iters[global_id] = temp_r + a[global_id];
        b_iters[global_id] = temp_i + b[global_id];

      }

      if (a_iters[global_id]*a_iters[global_id] + b_iters[global_id]*b_iters[global_id] >=4)
        {
          break;
        }
  }
  if (a_iters[global_id]*a_iters[global_id] + b_iters[global_id]*b_iters[global_id] <=4)
  {
       result_iters[global_id] = 1;
  }
}
"""
).build();
```

# 3: Generando c

Como sabemos la función $\texttt{numpy.random.rand(n)}$ genera un vector de tamaño *n* con valores aleatorios entre 0 y 1, de modo que los reescalamos con la función $g(x) = 4x -2$, veamos que es creciente y que f(0) = -2 y f(1) = 2.

Por otro lado no se puede paralelizar la genración de puntos aleatorios pues la función es pseudorandom, es decir toma información sobre tiempo para generar aleatoriamente un número por lo que si decidimos paralelizarlo pueden generarse puntos duplicados.

# 4: Calculando!

Comenzaremos con un ejemplo secuencial para tener una idea del orden de tiempo que demoraría con, por ejemplo 10 mil puntos y 10 mil iteraciones.

```python
counter = 0
n = 10000
a = np.random.rand(n).astype(np.float32)*4-2
b = np.random.rand(n).astype(np.float32)*4-2

t0_np = tm.time()
for num in range(n):
    temp_r = a[num]
    temp_i = b[num]
    bandera = True
    for i in range(10000):
        t1 = temp_r ** 2 -temp_i**2
        t2 = 2*temp_r*temp_i
        temp_r = t1 + a[num]
        temp_i = t2 + b[num]
        if temp_r**2 + temp_i **2 >= 4:
            bandera = False
            break
    if bandera:
        counter += 1
        
t1_np = tm.time()

print(f"Tiempo de computo: {t1_np-t0_np}\nResultado: {16*counter/n}")
```

A continuación mostraremos el desempeño para los 3 programas, motrando el valor resultante y tiempo de ejecución.

```python
n = 33554432 #= 32*1024*1024 quise poner un multiplo del tamaño de los grupos que hice
m = 1024

a_h = np.random.rand(n).astype(np.float32)*4-2
b_h = np.random.rand(n).astype(np.float32)*4-2
result = np.zeros(n).astype(np.float32)


a = cl.Buffer(context, cl.mem_flags.READ_ONLY | cl.mem_flags.COPY_HOST_PTR, hostbuf=a_h)
b = cl.Buffer(context, cl.mem_flags.READ_ONLY | cl.mem_flags.COPY_HOST_PTR, hostbuf=b_h)

a_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, a_h.nbytes)
b_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, b_h.nbytes)

result_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, b_h.nbytes)


t0_program_iters = tm.time()
program.program_iters(queue, (n,), (m,), a, b, a_iters, b_iters,result_iters)
cl.enqueue_copy(queue, result, result_iters)
valor = 16*sum(result)/n
t1_program_iters = tm.time()
print("\nValor Only Iters: {} Tiempo de ejecucion: {}".format(valor, t1_program_iters-t0_program_iters))

t0_program_if_statement = tm.time()
program.program_if_statement(queue, (n,), (m,), a, b, a_iters, b_iters,result_iters)
cl.enqueue_copy(queue, result, result_iters)
valor = 16*sum(result)/n
t1_program_if_statement = tm.time()
print("\nValor If Statement: {} Tiempo de ejecucion: {}".format(valor, t1_program_if_statement-t0_program_if_statement))



t0_program_long_iters = tm.time()
program.program_long_iters(queue, (n,), (m,), a, b, a_iters, b_iters,result_iters)
cl.enqueue_copy(queue, result, result_iters)
valor = 16*sum(result)/n
t1_program_long_iters = tm.time()
print("\nValor Long Iters: {} Tiempo de ejecucion: {}".format(valor, t1_program_long_iters-t0_program_long_iters))
```

Veamos que los 3 progrmaas llegan al mismo resultado, lo que era esperable ya que usamos el mismo conjunto de puntos originales. Pero lo relevante es la cantidad de puntos utilizados y el tiempo de ejecución. Mientras que el tiempo de ejecución *Only Iters* es ligeramente mayor que el de *Long Iters* la razón por la que creo que pasa esto es porque deben existir muchos puntos que se escapan rapidamente y si en particular es uno de estos puntos el último del grupo en calcular atrasa a todo el programa. Por otro lado *If Statement* es el que demora más tiempo, alrededor de un 8% más que *Long Iters*, es menos de los esperado pero aún así una diferencia significativa. Finalmente se ve claramente una mejora de eficiencia mientras que ahora se usan alrededor de 33 millones de puntos con la misma cantidad de iteraciones, se demora el mismo orden de tiempo, si vemos el tiempo necesario por punto resulta:

  1. **GPU**: 0,000000674 segundos *(Long Iters)*
  2. **Secuencial**: 0,002156401 segundos

Esto significa que en el tiempo de calculo de un punto secuencial la GPU hace 3199,40 puntos.

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)
for i in range(len(result[0:50000])): #Dado que matplotlib no es lo mejor para hacer esto, solo lo haré con las primeras 50000 entradas.
    if result[i]:
        ax.plot(a_h[i],b_h[i], color='black', marker='o')
#un grafico bonito
fig.show()
```

```python
area_real = 1.50659177
i = 4
counters_list=[sum(result[0:2])]
error_list = [(area_real-counters_list[-1])/area_real]
while i<= 33554432:
    counters_list.append(counters_list[-1]+sum(result[int(i/2):i]))
    error_list.append(abs((area_real-16*counters_list[-1]/i)/area_real))
    i = i*2
```

```python
plt.figure(figsize=(20,20))
plt.loglog([2**i for i in range(1,26)],error_list)
#Ccantidad de puntos vs Error en loglog
```

# 5: Reducir el error

Podemos ver como se va reduciendo el error a medida que aumentamos las iteraciones, se podría disminuir aun mas si se llega a un punto en el cual hacemos con aún mas puntos (como veo que el error sigue disminuyendo sin importar aún se pueden agregar más puntos para que se reduzca)

```python
lista_valores = [valor];
```

```python
for i in range(10):
    a_h = np.random.rand(n).astype(np.float32)*4-2
    b_h = np.random.rand(n).astype(np.float32)*4-2
    result = np.zeros(n).astype(np.float32)


    a = cl.Buffer(context, cl.mem_flags.READ_ONLY | cl.mem_flags.COPY_HOST_PTR, hostbuf=a_h)
    b = cl.Buffer(context, cl.mem_flags.READ_ONLY | cl.mem_flags.COPY_HOST_PTR, hostbuf=b_h)

    a_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, a_h.nbytes)
    b_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, b_h.nbytes)

    result_iters = cl.Buffer(context, cl.mem_flags.WRITE_ONLY, b_h.nbytes)

    t0_program_long_iters = tm.time()
    program.program_long_iters(queue, (n,), (m,), a, b, a_iters, b_iters,result_iters)
    cl.enqueue_copy(queue, result, result_iters)
    valor = 16*sum(result)/n
    t1_program_long_iters = tm.time()
    print("\nValor Long Iters: {} Tiempo de ejecucion: {}".format(valor, t1_program_long_iters-t0_program_long_iters))
    lista_valores.append(valor)
```

```python
print(f"Aproximación final: {sum(lista_valores)/len(lista_valores)}")
print(f"Error anterior: {error_list[-1]}")
print(f"Error final: {(sum(lista_valores)/len(lista_valores)-area_real)/area_real}")
```

Finalmente, creo que si llegamos al punto en el cual son necesarias más iteraciones para tener una mejor aproximación del error, pues habiendo calculado aún más puntos (10 veces más) tuvimos un empeoramiento del resultado, por lo que sería necesario ser más precisos al determinar si un punto está dentro o no del fractal de Mandelbrot
